```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load Image
image = cv2.imread('./lenna.png')

# Convert BGR format to RGB format
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 1. Apply Gaussian Filter 
gaussian_filtered = cv2.GaussianBlur(image_rgb, (15, 15), 0)

# 2. Apply Bilateral Filter
bilateral_filtered = cv2.bilateralFilter(image_rgb, 15, 75, 75)

# 3. Apply Non-Local Means Filter
non_local_means_filtered = cv2.fastNlMeansDenoisingColored(image_rgb, None, 10, 10, 7, 21)

# Visualize the results
fig, axs = plt.subplots(1, 4, figsize=(20, 10))
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(gaussian_filtered)
axs[1].set_title('Gaussian Filter')
axs[1].axis('off')

axs[2].imshow(bilateral_filtered)
axs[2].set_title('Bilateral Filter')
axs[2].axis('off')

axs[3].imshow(non_local_means_filtered)
axs[3].set_title('Non-Local Means Filter')
axs[3].axis('off')

plt.tight_layout()
plt.show()
```
